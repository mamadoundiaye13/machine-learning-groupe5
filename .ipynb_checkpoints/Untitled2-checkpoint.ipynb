### Nettoyage des donn√©es

```python
# les valeurs null
```

```python
print('Nombre de valeurs null par colonne:',clients.isnull().sum(), sep = '\n')
```

```python
for cols in ['Administrative','Informational','ProductRelated']:
    print('le nombre de valeur nominale de la colonne {} :'.format(cols),clients[cols].isnull().sum(), sep = '\n')
```

```python
#les colonnes avec des donn√©es nominales 'Administrative','Informational','ProductRelated'
```

```python
for cols in ['Administrative','Informational','ProductRelated']:
    clients[cols].replace(0, np.nan, inplace= True)
    
    
for cols in ['Administrative','Informational','ProductRelated']:
    print('le nombre de valeur nominale de la colonne {} :'.format(cols),clients[cols].isnull().sum(), sep = '\n')
```

```python
# imputation avec la mediane 
```

```python
for cols in ['Administrative','Informational','ProductRelated']:
    median_value=clients[cols].median()
    clients[cols]=clients[cols].fillna(median_value)
    
    
for cols in ['Administrative','Informational','ProductRelated']:
    print('le nombre de valeur null de la colonne {} :'.format(cols),clients[cols].isnull().sum(), sep = '\n')
```

```python
#les colonnes avec des valeurs n√©gatives 'Administrative_Duration','Informational_Duration','ProductRelated_Duration','BounceRates','ExitRates'
```

```python
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration']:
    clients[cols].replace(0, np.nan, inplace= True)
    
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration']:
    clients[cols].replace(-1, np.nan, inplace= True)
    
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration']:
    print(' Le nombre de valeur n√©gative de la colonne {}:'.format(cols),clients[cols].isnull().sum(), sep = '\n')
```

```python
#les imputer avec la moyenne
```

```python
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration','BounceRates','ExitRates']:
    mean_value=clients[cols].mean()
    clients[cols]=clients[cols].fillna(mean_value)
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration','BounceRates','ExitRates']:
    print('le nombre de valeur null de la colonne {}:'.format(cols),clients[cols].isnull().sum(), sep = '\n')
```

### Mise √† l'√©chelle

```python
Categorical_variables=['Weekend','Revenue','Administrative','Informational','ProductRelated','SpecialDay',
 'OperatingSystems','Browser','Region','Month','TrafficType','VisitorType']
```

```python
from sklearn.preprocessing import StandardScaler

scale=[feature for feature in clients.columns if feature not in Categorical_variables]

scaler=StandardScaler()
scaler.fit(clients[scale])
```

```python
scaled_data = pd.concat([clients[['Weekend','Revenue','Administrative','Informational',
                                    'ProductRelated','SpecialDay','OperatingSystems',
                                    'Browser','Region','Month','TrafficType','VisitorType']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(clients[scale]), columns=scale)],
                    axis=1)
scaled_data.head()
```

### Labelisation

```python
from sklearn.preprocessing import LabelEncoder
encoded_features=['Month','VisitorType']

label_data = scaled_data.copy()
label_encoder = LabelEncoder()
for col in encoded_features:
    label_data[col] = label_encoder.fit_transform(scaled_data[col])
    
label_data.head()
```

### Train et test 

```python
from sklearn.model_selection import train_test_split

X=label_data.drop(['SpecialDay','VisitorType','Weekend','Revenue'],axis=1)
y=label_data.Revenue

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8,random_state=1)
print("Input du Training:",X_train.shape)
print("Input du Test:",X_test.shape)
print("Output du Training:",y_train.shape)
print("Output du Test:",y_test.shape)
```
